```python
#set the header
olympics_df = pd.read_csv('olympics.csv', header = 1)
olympics_df.head()
```

```python
#rename the columns
new_names = {'Unnamed: 0' : 'Country', '? Summer' : 'Summer Olympics','01 !' : 'Gold', '02 !' : 'Silver', '03 !' : 'Bronze', '? Winter' : 'Winter Olympics', '01 !.1' : 'Gold_1', '02 !.1' : 'Silver_1', '03 !.1' : 'Bronze_1','Total.1': 'Total_1', '? Games' : '# Games', '01 !.2' : 'Gold_2', '02 !.2' : 'Silver_2', '03 !.2' : 'Bronze_2', 'Combined total': 'Combined_total'}
olympics_df.rename(columns = new_names, inplace = True)
olympics_df.head()
```

```python
#sort the data by a specific column
o_df = olympics_df.sort_values(by = 'Combined_total', ascending = False)
o_df.head(10)
```

```python
#check combined total ?=  Total + Total.1
import engarde.decorators as ed
def addition(olympics_df):
    a = (olympics_df.Total + olympics_df.Total_1 == olympics_df.Combined_total)
    return a

@ed.is_shape((-1, 16))
@ed.verify_all(addition)
def unload():
    medals = olympics_df
    return medals
```

```python
df = unload()
```
